```python
# Read Data Set in specified Format
chiporder_df = pd.read_table('http://bit.ly/chiporders') # tab seperated file
```

```python
chiporder_df.head(3)
#Show the top 3 row of data table
```

```python
#describing a numerical series
print(chiporder_df.describe() )

#if data is not numerical then describe shows the categorial description
#like
r_data = pd.Series(['a', 'a', 'b', 'c'])
print("\n")
print(r_data.describe() )
```

```python
chiporder_df.dtypes
#show the data types of all columns of df
```

```python
chiporder_df.item_price = chiporder_df.item_price.str.replace("$",'').astype('float')
#clean $ sign from item_price and convert into float format known as data munging
```

```python
# read movie users data 
movieusers__df = pd.read_table("http://bit.ly/movieusers",sep="|", header=None)
```

```python
movieusers__df
```

```python
# Create A list for columns name
col_names = ["ID", "Age", "Gender", "Occupation", "Revenue"]
```

```python
#replace the columns name with movieusers__df
movieusers__df.columns = col_names
```

```python
movieusers__df.set_index("ID", inplace=True)
```
